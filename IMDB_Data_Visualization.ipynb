```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
'''
[2001-2005] Action;Thriller 274
[2001-2005] Adventure;Sci-Fi 44
[2001-2005] Comedy;Romance 1077 
[2006-2010] Action;Thriller 398
[2006-2010] Adventure;Sci-Fi 73
[2006-2010] Comedy;Romance 1334
[2011-2015] Action;Thriller 715
[2011-2015] Adventure;Sci-Fi 167
[2011-2015] Comedy;Romance 1953

'''

new_df = pd.DataFrame({'YearRange':['2001-2005','2006-2010','2011-2015'],
                       'ActionThriller':pd.Series([274,398,715]),
                       'ActionSci-Fi': pd.Series([44,73,167]),
                       'ComedyRomance': pd.Series([1077,1334,1953])
                       
                      })
new_df
```

```python
#plot of Comedy;Romance with Year

plt.plot(new_df.YearRange,new_df["ActionThriller"],label="Action Thriller")
plt.plot(new_df.YearRange,new_df["ActionSci-Fi"],label="Action Sci-Fi")
plt.plot(new_df.YearRange,new_df["ComedyRomance"],label="Comedy Romance")
plt.xlabel("Year Range")
plt.ylabel("Number of  Movies")
plt.legend(loc='upper left');
```

```python
#StackPlot

labels=["Action Thriller","Action Sci-Fi","Comedy Romance"]

y = np.stack([new_df["ActionThriller"].astype(np.int32),new_df["ActionSci-Fi"].astype(np.int32),new_df["ComedyRomance"].astype(np.int32)])

fig,ax = plt.subplots()
ax.stackplot(new_df.YearRange,new_df["ActionThriller"].astype(np.int32),new_df["ActionSci-Fi"].astype(np.int32),new_df["ComedyRomance"].astype(np.int32),labels=labels)
ax.legend(loc='upper left')
plt.xlabel("Year Range")
plt.ylabel("Number of  Movies")
plt.show();
```

```python
#Bar plot
plt.bar(new_df.YearRange, new_df["ComedyRomance"],label="Comedy Romance")
plt.bar(new_df.YearRange, new_df["ActionThriller"],label="Action Thriller")
plt.bar(new_df.YearRange, new_df["ActionSci-Fi"],label="Action Sci-Fi")
plt.xlabel("Year Range")
plt.ylabel("Movies Count in each genre")
plt.title("Distribution of each genre movies")
plt.legend(loc='upper left');
```
